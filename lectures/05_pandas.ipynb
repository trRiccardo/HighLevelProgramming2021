# Pandas

The `numpy` library is excellent for numerical computations, but it lacks support to <mark>handle missing data or non-omogeneous arrays</mark>. The `pandas` library is <mark>based on numpy and extends the numpy functionality</mark>, and is currently one of the most widely used tools for data manipulation, providing high-performance, easy-to-use data structures and advanced data analysis tools.

In particular `pandas` features:

* A fast and efficient `DataFrame` object for <mark>data manipulation</mark> with integrated indexing;
* Tools for reading and writing data between in-memory data structures and <mark>different formats (CSV, Excel, SQL, HDF5)</mark>;
* Convenient label-based slicing, fancy indexing, and subsetting of <mark>large data sets</mark>;
* Hierarchical axis indexing provides an intuitive way of working with <mark>high-dimensional data in a lower-dimensional data structure</mark>;
* Smart data alignment and integrated <mark>handling of missing data</mark>;
* <mark>Aggregating and transforming data with a powerful "group-by" engine</mark>; 
* <mark>High performance merging and joining of data sets</mark>;
* <mark>Time series-functionalities</mark>;
* Highly optimized for <mark>performance</mark>, with critical code paths written in Cython or C.


````python
import pandas as pd # standard naming convention
import numpy as np
````

## Series

Pandas Series represent an <mark>extension of the numpy 1D arrays</mark>. A Series is equivalent to a numpy array, but the <mark>axis is labeled</mark>, and there is the possibility to store <mark>heterogeneous data</mark>. Labels doesn't need to be unique but must be a hashable type.

One of the most important examples are the <mark>time-series</mark>, which are used to keep track of the time evolution of a certain quantity.

Link to the official Pandas Series [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html).

````python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
    
# calling the pd.Series(...) constructor
sr = pd.Series(np.arange(10)*0.5, index=letters[:10], dtype=float) # data, index (optional), type (optional)

# print series
print("Series:\n", sr)

# print indeces of data
print("\nIndices:\n", sr.index)

# values of the Series are a numpy array
print("\nValues:", sr.values, type(sr.values))

# print data type
print("\nType:\n", sr.dtype)

# accessing elements
print("\nElement by index     :", sr['f']) # like arrays
print("\nElement by attribute :", sr.f) # like attributes (not recommended)
subsr = sr[['d', 'f', 'h']] # note the double square brackets
print("\nSeries subset:\n", subsr, type(subsr)) # multiple indexing returns another series
````

````python
# Extracting elements and operations are the same as numpy array
print(sr[:3], '\n')
print(sr[7:], '\n')
print(sr[::3], '\n')
print(sr[sr > 3], '\n')
print(np.exp(sr), '\n')
print(np.mean(sr), np.std(sr), '\n')
````

Series may contain <mark>non-omogeneous data; in this case, the data type is referred to as `object`.</mark> Non-homogeneous data is normally handeled also by pandas and does not represent a problem, however this pays the price of <mark>less time-efficient operations.</mark>

````python
# Series can be created from a python dictionary, too
# Note that the elements can be of different types
d = {'b' : 1, 'a' : 'cat', 'c' : [2, 3]}
so = pd.Series(d)
print(so, '\n')
````

A key difference between pandas Series and numpy arrays is that <mark>operations between Series **automatically align the data based on the label**</mark>.

Thus, you can write operations without considering whether the Series involved have the same labels, or even the same size.

````python
# data
s = pd.Series(np.arange(5), index=tuple(letters[:5]))
print("Series:\n", s)

# sum elements by shifting the series
s1 = s[1:] + s
print("\nShifted sum:\n", s1)

s2 = s[1:] + s[:-1]
print("\nDouble shifted sum:\n", s2)
````

### Time series

Time series are very often used to describe the <mark>behaviour of a quantity as a function of time.</mark> Pandas has a special type of <mark>index for that, `DatetimeIndex`, that can be created e.g. with the function `pd.data_range()`.</mark>

````python
# to define a date, the datetime module is very useful
import datetime as dt

date = dt.date.today()
print("Today's date:\n", date)

# specify year, month, day, hour, minutes, seconds, and us
date = dt.datetime(2020, 11, 12, 10, 45, 10, 15)
print("\nDate and time:\n", date)

# otherwise, several notations can also be used
date = 'Nov 9 2020'
# or alternatively
date = '9/11/2020 14:45:00'
print("\nDate format:\n", date)

# create DatetimeIndex using ranges:
days = pd.date_range(date, periods=7, freq='D') # starting date, periods=number of dates, freq=time step
print("\n7 days range:\n", days)

seconds = pd.date_range(date, periods=3600, freq='s')
print("\n1 hour in seconds:\n", seconds)
````

To learn more about the frequency strings, please check the [documentation](http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases).

<mark>Timestamped data</mark> is the most basic type of time series data that associates values with points in time.

Functions like `pd.to_datetime` can be used to <mark>convert between different formats</mark> and, <mark>for instance, when reading the time stored as a string from a dataset:</mark>

````python
# Get the timestamp, which is the nanoseconds from January 1st 1970
tstamp = pd.Timestamp(date)
#tstamp = pd.Timestamp(dt.datetime(1970, 1, 1, 0, 0, 0, 1))
print("Timestamp:\n", tstamp.value)

# when creating a timestamp the format can be explicitly passed
ts = pd.to_datetime('2010/11/12', format='%Y/%m/%d')
print("\nTime:", ts, ", timestamp:", ts.value, ", type:", type(ts))

ts = pd.to_datetime('12-11-2010 10:39', format='%d-%m-%Y %H:%M')
print("\nTime:", ts, ", timestamp:", ts.value, ", type:", type(ts))
````

A standard series can be created, and (a range of) elements can be used as indices:

````python
print("index:\n", days, '\n')
tseries = pd.Series(np.arange(len(days)), index=days)
print("time series:\n", tseries, '\n')
# Extracting elements
print("slice by position:\n", tseries[0:4], '\n')
print("slice by date range:\n", tseries['2020-9-11' : '2020-9-14'], '\n') # note that includes end time
````

`pd.to_datetime` can also be used to create a `DatetimeIndex` if the argument is a list:

````python
print(pd.to_datetime([1, 2, 3, 4], unit='D', origin=pd.Timestamp('1980-02-03')))
````

## DataFrame

A pandas DataFrame can be thought as a <mark>tabular spreadsheet</mark>, although the performance, the functionalities and the capabilities are very different.

Similarly to Series, the DataFrame structure also contains <mark>labeled axes (rows and columns).</mark> <mark>Arithmetic operations **align on both row and column labels**.</mark> <mark>Each column in a DataFrame is a Series object</mark>: as a matter of fact, a DataFrame can be thought of as a dict-like container for Series objects.

The elements can be of all types, and <mark>missing data</mark> could be present too (represented as NaN).

For future reference (or for people already familiar with R), a pandas DataFrame is also similar to the R DataFrame.

Link to the official [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html).

### DataFrame constructor

A DataFrame objects can be created by <mark>passing a dictionary of objects</mark>. Note that the dictionary <mark>values are not omogeneous and do not have the same length.</mark> In these cases, pandas will <mark>automatically adjust the sizes</mark>, by replicating the content or adding NaN if necessary.

````python
df = pd.DataFrame({
    'A' : 1.,
    'B' : pd.Timestamp('20130102'),
    'C' : pd.Series(3, index=range(4), dtype='float32'),
    'D' : np.arange(7, 11),
    'E' : pd.Categorical(["test", "train", "test", "train"]),
})
df
````

An example of DataFrame with a DatatimeIndex object as index:

````python
entries = 10
columns = ['A', 'B', 'C', 'D']
dates = pd.date_range('11/9/2020 14:45:00', freq='h', periods=entries) # days/month/year
df = pd.DataFrame(np.arange(entries*4).reshape(entries, len(columns)), index=dates, columns=columns)
df # pay attention that the date is printed as year-day-month
````

## Selection

### Slicing

DataFrame <mark>slicing allows to select a subset of the DataFrame, or an entire column (a Series):</mark>

````python
## standard and safe
print(df['A'], '\n', type(df['A']), '\n') # Returns a Series (a column)

## equivalent but dangerous (imagine blank spaces in the name of the column, or a column named "T")
print(df.A, '\n')
````

````python
# selecting rows by position. Returns another DataFrame (a copy)
print(df[0:3])

# or by index range
print(df["2020-11-09 14:45:00" : "2020-11-09 16:45:00"])
````

### Selection by label

The most common way to <mark>select elements, rows, or columns in a DataFrame is by using the `.loc[]` method.</mark>

`.loc` supports <mark>multi-indexing, and returns a **copy** of the DataFrame (not a view, `.iloc[]` creates a **view** but has a different sintax).</mark>

````python
# getting a part of the DataFrame (in this case, a row)) using a label. Returns a Series
dfs = df.loc[dates[0]]
print(dfs, '\n', type(dfs), '\n')
````

````python
# selecting on a multi-axis by label:
dfa = df.loc[:, ['A','B']]
print("Are df and dfa the same object?", np.may_share_memory(df, dfa)) # -> the second is a copy, another object
dfa
````

````python
# showing label slicing, both endpoints are included:
df.loc['2020-11-09 18:45:00':'2020-11-09 20:45:00', ['A','B']]
````

````python
# getting an individual element
print(df.loc[dates[1], 'A'], '\n', type(df.loc[dates[1], 'A']), '\n')
````

The <mark>`.at()` method is equivalent</mark> to `.loc[]`. Use `at` if you only need to <mark>get or set a single value</mark> in a DataFrame or Series.

````python
print(df.at[dates[1], 'A'])
````

### Selecting by position

`.iloc[]` is similar ot `.loc[]`, but instead of labels, <mark>it uses pure integer-location based indexing for selection by position.</mark>

But differently from `.loc[]`, `.iloc[]` <mark>returns a **view**, not a copy.</mark>

Yes, views and copies are ambiguous in DataFrames. And it gets worse! This appears to be a long standing issue for pandas. Read a nice article on this topic [here](https://www.practicaldatascience.org/html/views_and_copies_in_pandas.html).

````python
# select via the position of the passed integers:
print(df.iloc[3], '\n')

# row and column ranges selected with numpy-like notation:
dfv = df.iloc[3:5, 0:2]
print(dfv, '\n')

print("Are df and dfv the same object?", np.may_share_memory(df, dfv))
````

````python
# selecting rows 1,2 and 4 for columns 0 and 2
df.iloc[[1, 2, 4], [0, 2]]
````

````python
# slicing rows explicitly
df.iloc[1:3, :]

# slicing columns explicitly
df.iloc[:, 1:3]
````

Similary to `.loc[]` and `.at[]`, there is <mark>also `.iat[]` alongside `.iloc[]`:</mark>

````python
# selecting an individual element by position: no difference between iloc and iat
print(df.iloc[1,1], ", type:", type(df.iloc[1,1]))
print(df.iat[1,1], ", type:", type(df.iat[1,1]))
````

### Masks

<mark>Boolean masks can be used in the same way as numpy</mark>, and they represent a very powerful way of filtering out data with certain features. Just like numpy fancy indexing, using a mask <mark>returns a **copy** of the DataFrame.</mark>

````python
# Selecting on the basis of boolean conditions applied to the whole DataFrame
dfc = df[df > 10]
dfc.iat[0, 0] = -999
# a DataFrame with the same shape is returned, with NaN's where condition is not met
# Note that when a NaN is present in a column of integers, the column (Series) is casted to float
print("Are df and dfc the same object?", np.may_share_memory(df, dfc))
dfc
````

````python
# Filter by a boolean condition on the values of a single column
dfc[dfc['B'] < 20.]
````

### Assignement

<mark>Assignment is typically performed after selection:</mark>

````python
# Make sure to copy the DataFrame if you plan to modify it, and you don't want to change the original object
dfa = df.copy()

# setting values by label (same as by position)
dfa.at[dates[0], 'A'] = -1

# setting and assigning a numpy array
dfa.loc[:, 'D'] = np.array([5] * len(dfa))

# defining a new column
dfa['E'] = np.arange(len(dfa))*0.5

# defining a brend new column by means of a pd.Series: indexes must be the same!
dfa['E prime'] = pd.Series(np.arange(len(dfa))*2, index=dfa.index)

# using masks for assigment
dfa[dfa >= 30] = -dfa

dfa
````

### Application of a function

User-defined or standard <mark>functions can be applied on entire DataFrames or columns, with very short execution times:</mark>

````python
def dcos(theta):
    theta = theta*(np.pi/180)
    return np.cos(theta)
 
dfa['cosine'] = dfa["E"].apply(dcos)
dfa
````

### Dropping

<mark>Dropping columns is an example of a method that does not modify the original object, and returns a new modified object.</mark> In other words, if you want to <mark>keep the modified DataFrame, perform a new assignment:</mark>

```python
df = df.drop(....)
```
Alternatively, the <mark>modification of the original object can be forced by specifying `inplace=True`</mark> among the arguments.

````python
dfb = dfa.copy()

# Dropping by column: it returns the modified object; I have to assign it to something...
dfb.drop(['E prime'], axis=1)

#which is equivalent to
dfb = dfb.drop(columns=['E prime'])
#dfb.drop(columns=['E prime'], inplace=True)

dfb
````

## Missing data

Pandas primarily uses the <mark>value `np.nan` to represent missing data.</mark> It is by default <mark>not included in computations.</mark> If there is a NaN entry in a Series of integers, the <mark>type of the Series will be changed to floats.</mark>

````python
df_wNan = dfb[dfb > 0]
df_wNan
````

````python
# dropping rows with at least a Nan
df_wNan.dropna(how='any')
````

````python
# getting a mask
df_wNan.isna()
#df_wNan.notna()
````

````python
# filling missing data (not recommended, unless you really mean it)
df_wNan.fillna(value=0)
````

## Merge

Pandas provides various functions for easily <mark>combining together Series and DataFrames in join / merge-type operations.</mark>

### Concat

<mark>Concatenation (adding rows)</mark> is straightforward:

````python
rdf = pd.DataFrame(np.arange(40).reshape(10, 4))
rdf
````

````python
# split DataFrame into 3 pieces, row-wise
pieces = [rdf[:3], rdf[3:7], rdf[7:]]
pieces
````

````python
# put it back together
pd.concat(pieces)

# in this case, indices are already set; if they are not, indices can be ignored
#pd.concat(pieces, ignore_index=True)
````

In case of <mark>dimension mismatch, Nan are added where needed.</mark>

<mark>Appending rows and columns also works:</mark>

````python
# appending a single row (as a Series)
s = rdf.iloc[3]
rdf = rdf.append(s, ignore_index=True) # remember to assign the returned object, or use inplace=True
rdf
````

### Merge/Join

<mark>SQL like operations</mark> on table can be performed on DataFrames. This is a quite advanced use case, refer to the [doc](https://pandas.pydata.org/pandas-docs/stable/merging.html#merging) for more info/examples.

````python
left = pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'bar'], 'rval': [4, 5]})

pd.merge(left, right, on="key")
````

## Grouping

In real world applications, it's quite common that <mark>several entries (row) belong to a certain entity, or "group".</mark> DataFrames have a powerful tool to perform operations on entries of the same group. The method is called <mark>`.groupby()`, and it usually involves one or more of the following steps:</mark>

* <mark>Splitting the data into groups based on some criteria</mark>
* <mark>Applying a function to each group independently</mark>
* <mark>Combining the results into a data structure</mark>


````python
gdf = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                    'B' : ['one', 'one', 'two', 'three',
                           'two', 'two', 'one', 'three'],
                    'C' : np.arange(8),
                    'D' : np.linspace(10, -10, 8)})
gdf
````

````python
# Grouping and then applying the sum() 
# function to the resulting groups (effective only where numerical values are present)
print(gdf.groupby('A'))
gdf.groupby('A').sum()
````

````python
# Example: find maximum value in column D for each group, and assign the value to a new column
gdf['M'] = gdf.groupby('A')['D'].transform(np.max)
gdf
````

## Multi-indexing

<mark>Hierarchical / Multi-level indexing</mark> allows sophisticated data analysis on <mark>higher dimensional data.</mark> In practice, it enables you to <mark>store and manipulate data with an arbitrary number of dimensions in lower dimensional data structures like Series (1D) and DataFrames (2D).</mark>

````python
# Creat multi-dimensional index
tuples = list(zip(['bar', 'bar', 'baz', 'baz', 'foo', 'foo', 'qux', 'qux'],
          ['one', 'two', 'one', 'two', 'one', 'two', 'one', 'two']))
multi_index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
print(multi_index, '\n', type(multi_index), '\n')

# Create multi-indexed dataframe or series
s = pd.Series(np.arange(8)/np.pi, index=multi_index)
s
````

````python
# multi-indexing enables further features of the groupby method,
# e.g. when group-by by multiple columns
gdf.groupby(['A','B']).sum()
````

````python
itime = dt.datetime.now()
print("Begin time:", itime)

# the one-liner command
data['TIMENS'] = data['TDC_MEAS'] * 25 / 30 + data['BX_COUNTER'] * 25

ftime = dt.datetime.now()
print("End time:", ftime)
print("Elapsed time:", ftime - itime)

data
````

````python
# the loop
def conversion(data):
    result = []
    for i in range(len(data)): 
        result.append(data.loc[data.index[i], 'TDC_MEAS'] * 25 / 30. + data.loc[data.index[i], 'BX_COUNTER'] * 25)
    return result

itime = dt.datetime.now()
print("Begin time:", itime)
data['TIMENS'] = conversion(data)
ftime = dt.datetime.now()
print("End time:", ftime)
print("Elapsed time:", ftime - itime)

data
````
